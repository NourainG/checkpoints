```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
df = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
df
```

```python
#importing librairies 
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
#we apply the one hot encoding on the df
te_ary=te.fit(df).transform(df)
#create a dataframe from a numy array
df=pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
#select itemsets
from mlxtend.frequent_patterns import apriori 
apriori(df,min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
#extract rules by other metrics for exemple confidence
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.6)
```

```python
#using the metric "lift"
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="lift",min_threshold=1.1)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
df_new=pd.read_csv("C:/file/Market_Basket_Optimisation.csv",sep=',')
df_new
```

```python
df_new.isnull().sum()
```

```python
df_new.shape
df_new.dtypes
```

```python
#replacing the missing alues with the frequent ones:
df_new["avocado"].fillna("mineral water",inplace=True)
df_new["vegetables mix"].fillna("mineral water",inplace=True)
df_new["green grapes"].fillna("green tea",inplace=True)
df_new["whole weat flour"].fillna("french fries",inplace=True)
df_new["yams"].fillna("green tea",inplace=True)
df_new["salad"].fillna("green tea",inplace=True)
df_new["low fat yogurt"].fillna("green tea",inplace=True)
df_new.drop(['salmon','antioxydant juice','frozen smoothie','spinach','olive oil'],1,inplace=True)
```

```python
#as pre-processing step we can convert the values of the dataframe into an array of strings representing transactions 
import numpy as np
trans=[]
for i in range(0,7500):
    trans.append([str(df_new.values[i,j]) for j in range (0,15)])
trans= np.array(trans)
print(trans.shape)
    
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
#we apply the one hot encoding on the df
df_new=te.fit_transform(trans)
#create a dataframe from a numy array
df_new=pd.DataFrame(df_new, columns = te.columns_)
df_new.head(5)
```

```python
# Association Rule Mining on the input dataframe
frequent_itemsets= apriori(df_new,min_support=0.8, use_colnames=True)
frequent_itemsets['length']= frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.8)
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="lift",min_threshold=0.8)
```
